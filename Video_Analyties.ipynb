```python
def video_to_image(path, folder):
    global exp_fld
    
    #importing video
    vidcap = cv2.VideoCapture(path)
    exp_fld = folder
    
    try:
        if not os.path.exists(exp_fld):
            os.makedirs(exp_fld)
    except OSError:
        print('Error: Creating directory of data')
        
    Count = 0
    sec = 0
    frameRate = 1
    while(True):
        vidcap.set(cv2.CAP_PROP_POS_MSEC, sec*1000)
        hasFrames, image = vidcap.read()
        sec = sec + frameRate
        sec = round(sec,2)
        
        #Exporting the image
        if hasFrames:
            name='./' + exp_fld + '/frame' + str(Count) + '.jpg'
            cv2.imwrite(name, image)
            
            Count +=1
        else:
            break
            
    return print("Image Exported")
```

```python
import cv2
import os

def video_to_image(path, folder):
    # Create target folder if it does not exist
    if not os.path.exists(folder):
        try:
            os.makedirs(folder)
        except OSError as e:
            print('Error: Creating directory of data', e)
            return

    # Load video
    vidcap = cv2.VideoCapture(path)
    if not vidcap.isOpened():
        print("Error: Couldn't open video file.")
        return

    count = 0
    sec = 0
    frameRate = 1  # Adjusts the frame rate for extracting images

    while True:
        vidcap.set(cv2.CAP_PROP_POS_MSEC, sec*1000)
        hasFrames, image = vidcap.read()
        sec = sec + frameRate
        sec = round(sec, 2)

        if hasFrames:
            name = f'./{folder}/frame{count}.jpg'
            cv2.imwrite(name, image)
            print(f'Exporting frame {count} at second {sec}')
            count += 1
        else:
            break

    print(f'{count} images exported')
    vidcap.release()

video_to_image("retail_store.mp4", 'crowd')
```

### Data Preparation

```python
#Resize the image
def img_re_sizing(dnst_mp, image):
    #Normalizing
    dnst_mp = 255 * dnst_mp / np.max(dnst_mp)
    dnst_mp = dnst_mp[0][0]
    image = image[0]
    #empty image 
    result_img = np.zeros((dnst_mp.shape[0]*2, dnst_mp.shape[1]*2))
    
    #iterate for each image
    for i in range(result_img.shape[0]):
        for j in range(result_img.shape[1]):
            result_img[i][j] = dnst_mp[int(i / 2)][int(j / 2)] /4
            
    result_img = result_img.astype(np.uint8, copy=False)
    
    return result_img
```

```python
def generate_dstys_map(o, dsty, cc, image_location):
    #define the fgr_imure
    fgr_im = plt.figure()
    
    #define size
    col = 2
    rws = 1
    X = o
    
    add = int(np.sum(dsty))
    dsty= img_re_sizing(dsty, o)
    
    #adding original image and new generated heatmap image
    for i in range(1, col*rws +1):
        if i == 1:
            image = X
            fgr_im.add_subplot(rws, col, i)
            #setting axis
            plt.gca().set_axis_off()
            plt.margins(0,0)
            #locator
            plt.gca().xaxis.set_major_locator(plt.NullLocator())
            plt.gca().yaxis.set_major_locator(plt.NullLocator())
            #adjust subplots
            plt.subplots_adjust(top=1, bottom = 0, right=1, left=0 ,hspace= 0, wspace=0)
            #show image
            plt.imshow(image)
            
        #generate dstys image
        if i == 2:
            image = dsty
            fgr_im.add_subplot(rws,col, i)
            #setting axis
            plt.gca().set_axis_off()
            plt.margins(0,0)
            #locator
            plt.gca().xaxis.set_major_locator(plt.NullLocator())
            plt.gca().yaxis.set_major_locator(plt.NullLocator())
            
            #adjust subplots
            plt.subplots_adjust(top = 1, bottom = 0, right = 1,left = 0, hspace = 0, wspace = 0)
             #adding count
            plt.text(1, 80, 'M-SegNet* Est: '+str(add)+',Gt:'+str(cc), fontsize=7, weight="bold",color = 'w')
            #show image
            plt.imshow(image)#, cmap=CM.jet)
            
    #image_nm with location
    #image_nm = image_location.split('/')[-1]
    #image_nm = image_nm.replace('.jpg', '_heatpmap.png')
    #image_nm = image_nm.replace('.jpg')
    
    base_dir, image_name = os.path.split(image_location)
    image_name = image_name.replace('.jpg', '_heatmap.png')
    save_dir = os.path.join(base_dir, 'seg_results')  # Specific subdirectory for saving results

    # Ensure the directory exists
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    save_path = os.path.join(save_dir, image_name)
    # Saving the image
    plt.savefig(save_path, transparent=True, bbox_inches='tight', pad_inches=0, dpi=200)
    plt.close(fgr_im)  # Close the figure to free up memory
     #saving the image
    #plt.savefig(image_location.split(image_nm)[0]+'seg_'+image_nm, transparent=True, bbox_inches='tight',pad_inches=0.0, dpi=200)
```

```python
#Getting count of peoples
def get_count_people(image):
    #simple preprocessing
    trans = transforms.Compose([transforms.ToTensor(),
                               transforms.Normalize([0.485,0.456, 0.406],
                                                   [0.229, 0.224, 0.225])])
    #sample image with height and width
    img = Image.open(image).convert("RGB")
    height, width = img.size[1], img.size[0]
    height = round(height / 16) * 16
    width = round(width / 16) * 16
    
    #re size the image
    img_den = cv2.resize(np.array(img), (width,height), cv2.INTER_CUBIC)
    
    #transform 
    img = trans(Image.fromarray(img_den))[None,:]
    model = M_SFANet_UCF_QNRF.Model()
    #define the model
    model.load_state_dict(torch.load('best_M-SFANet__UCF_QNRF.pth',map_location= torch.device('cpu')))
    
    model.eval()
    dnst_mp = model(img)
    #final count
    count = torch.sum(dnst_mp).item()
    
    #return count. density and map
    return count, img_den, dnst_mp
```

### Importing Images

```python
# Getting all the images from the path
image_location = []
path_sets = ['crowd']

for path in path_sets:
    for img_path in glob.glob(os.path.join(path,'*.jpg')):
        image_location.append(img_path)
        
image_location[:3]
```

### Getting Crowd Counts

```python
list_df = []
#loop through each image
for i in image_location:
    count, img_den, dnst_mp = get_count_people(i)
    generate_dstys_map(img_den, dnst_mp.cpu().detach().numpy(),0,i)
    list_df = list_df + [[i, count]]
    
#create the dataframe with image id and count
df = pd.DataFrame(list_df,column=['image','count'])

#sort and show
df.sort_values(['image']).head()
```
